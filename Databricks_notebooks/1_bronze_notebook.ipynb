###Data Overview

We have a total of 12 years (2010-2021) of data downloaded from the [UN Data repository](http://data.un.org/Data.aspx?d=UNHCR&f=indID%3AType-Ref) for this analysis. Given the size of the data set, I have filtered the data for the period desired directly from the source.
There is no license required for the use or publication of this data set.

### Create Database

Creating the **bronze** layer of or databases. It will store the main table with the raw data that we have loaded from a csv file in the DBFS.

```
%sql 

CREATE DATABASE mvp_bronze
```

###Create Table

Creating the table from the CSV file loaded in the DBFS. The columns were added correctly specifying the type of variable and an explanation of that variable.

```
%sql 

DROP TABLE bronze.un_refugees_raw
```

```
%sql

CREATE TABLE mvp_bronze.un_refugees_raw (
  country_to STRING COMMENT 'Country or territory of asylum or residence',
  country_from STRING COMMENT 'Country or territory of origin',
  year INT COMMENT 'Year of measurement',
  n_refugees INT COMMENT 'Number of refugees seeking asylum/residence',
  n_refugees_unhcr INT COMMENT 'Number os refugees assisted by UNHCR',
  total_refugee_like INT COMMENT 'Total refugees and people in refugee-like situations',
  total_refugee_like_unhcr INT COMMENT 'Total refugees and people in refugee-like situations assisted by UNHCR'
)
USING csv
LOCATION '/FileStore/tables/MVP/UNdata_Export_20240703_225347243.csv'
OPTIONS (HEADER 'true')
```
